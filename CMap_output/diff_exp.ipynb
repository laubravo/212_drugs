```r
library(BiocManager)
library(limma)
library(GEOquery)
library(dplyr)
library(tidyr)
```

```r
my_id <- "GSE141549"#"GSE141549"
gse <- getGEO(my_id)
gse <- gse[[2]]
exprs_gse <- read.delim('../data/GSE141549_data_rename.csv', header = TRUE, sep = ",", dec = " ")
 
exprs_gse
```

```r
m<-as.matrix(exprs_gse[,-1])
rownames(m)<-exprs_gse[,1]
mat_num <- matrix(as.numeric(m),    # Convert to numeric matrix
                  ncol = ncol(m))
  
mat_num # Print numeric matrix
```

```r
sampleInfo <- pData(gse)
sampleInfo <- rename(sampleInfo,group = "disease stage:ch1")
sampleInfo$group <- sampleInfo$group %>% replace_na('Healthy')
sampleInfo
#GSM4206272
```

```r
design <- model.matrix(~0+sampleInfo$group)
colnames(design) <- c("first","second", "third", "fourth", "healthy")
design
```

```r
fit <- lmFit(mat_num, design)
head(fit$coefficients)
```

```r
contrasts <- makeContrasts(healthy-(first+second+third+fourth), levels=design)

## can define multiple contrasts
## e.g. makeContrasts(Group1 - Group2, Group2 - Group3,....levels=design)

fit2 <- contrasts.fit(fit, contrasts)

fit2 <- eBayes(fit2)
fit2 <- topTable(fit2, number=28247)

tab <- as.data.frame(fit2)
write.table(tab, file='../differential_output/GSE141549_unstrat.csv', sep=',', append = FALSE)

```
